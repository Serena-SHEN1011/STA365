### Q2: Continue "Q2" of the previous weeks 3 and 4

1. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a normal distribution and your prior for $\tau$ is a gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

2. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a non normal distribution and your prior for $\tau$ is a non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

3. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a yet another different again non normal distribution and your prior for $\tau$ is a yet another different again non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

```python
import pymc as pm
import numpy as np
import arviz as az
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#Q2
# Set sample size and generate data
n = 30  # Number of observations
sample_mean = 0  # True mean
sample_variance = 1  # True standard deviation
x = stats.norm(loc=sample_mean, scale=sample_variance).rvs(size=n)  # Simulated data
```

```python
# 1. Bayesian inference with Normal prior for mean theta and Gamma prior for tau

with pm.Model() as model_1:
    # Priors
    theta = pm.Normal("theta", mu=0, sigma=10)  # Normal prior for mean
    tau = pm.Gamma("tau", alpha=1, beta=1)  # Gamma prior for precision
    
    # Likelihood
    y = pm.Normal("y", mu=theta, tau=tau, observed=x)
    
    # Inference
    trace_1 = pm.sample(2000, chains=2, tune=1000, return_inferencedata=True)
    
# Diagnostics
display(az.summary(trace_1))
az.plot_trace(trace_1)
plt.tight_layout()
```

```python
# 2. Bayesian inference with Non-Normal prior for theta and Non-Gamma prior for tau

with pm.Model() as model_2:
    # Priors
    theta = pm.Uniform("theta", lower=-10, upper=10)  # Uniform prior for theta
    tau = pm.Beta('tau', alpha=2, beta=2) # Beta prior for tau
    
    # Likelihood
    y = pm.Normal("y", mu=theta, tau=tau, observed=x)
    
    # Inference
    trace_2 = pm.sample(2000, chains=2, tune=1000, return_inferencedata=True)
    
# Diagnostics
display(az.summary(trace_2))
az.plot_trace(trace_2)
plt.tight_layout()
```

```python
# 3. Bayesian inference with another set of Non-Normal prior for theta and Non-Gamma prior for tau

with pm.Model() as model_3:
    # Priors
    theta = pm.Cauchy("theta", alpha=0, beta=2)  # Cauchy prior for theta
    tau = pm.Weibull("tau", alpha=2, beta=1)  # Weibull prior for tau
    
    # Likelihood
    y = pm.Normal("y", mu=theta, tau=tau, observed=x)
    
    # Inference
    trace_3 = pm.sample(2000, chains=2, tune=1000, return_inferencedata=True)
    
# Diagnostics
display(az.summary(trace_3))
az.plot_trace(trace_3)
plt.tight_layout()
```

### Q3: Slice Sampling

*First explain how the Markov algorithm of slice sampling as given below works. Then explain the steps by which slice sampling could be used in place of a Metropolis-Hasting step in a Metropolis within Gibbs algorithm where the full conditionals are only known up to a normalizing constant.
```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):
    
    # find interval of grid points where f(x_grid) > y 
    # then extend the enterval so endpoints f(a)<y and f(b)<y 
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid
    
    x_ = a + stats.uniform().rvs()*(b-a)    
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f" 
    elif x_ < x:  # or if "x_ was above f on the left side of the interval" 
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):
    
    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)
    
    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

**Overview of Slice Sampling**: Slice sampling is an adaptive Markov Chain Monte Carlo (MCMC) method used to sample from a target distribution $f(x)$ without needing to specify a proposal distribution. It works by introducing an auxiliary variable $y$ that defines a horizontal "slice" of the probability density function and then sampling uniformly from this region.

**Markov Algorithm in Slice Sampling**: 
1. Define the Target Function $f(x)$: The function $f(x)$ represents the probability density function (PDF) we want to sample from. 
2. Initialize a Grid for Sampling: Start with an initial value $x$ that lies within the support of the target distribution $f(x)$.
3. Find the Slice Interval `[0, f(x)]`, where $f(x) >y$: Identify the region where the function remains above $y$ and extend the interval slightly for numerical stability.
4. Propose a New Sample `x_` from `[0, f(x)]`: Sample uniformly from the interval `[0, f(x)]`.
5. Acceptance Check: If `f(x_) > y`, accept `x_` as the new sample. Otherwise, shrink the interval and repeat step 4.
6. Repeat Until Convergence: Iterate for a sufficiently large number of steps to ensure samples converge to the target distribution. `slice_f_at_y` and `slice_f_at_y_` functions implement this process. They iteratively shrink the interval until a suitable `x_` is found.

**Replacing Metropolis-Hastings in Gibbs Sampling**: 

Gibbs sampling is commonly used for sampling from high-dimensional distributions by iteratively sampling from each full conditional distribution. However, if a full conditional is only known up to a normalizing constant, direct sampling is difficult.

Reason for replacing Metropolis-Hastings in Gibbs Sampling: 
1. Avoids the Need for a Proposal Distribution: Metropolis-Hastings requires a carefully chosen proposal distribution, whereas slice sampling adapts to the target function automatically.
2. Metropolis-Hastings may reject many proposed samples, leading to slow mixing, whereas slice sampling directly samples from the conditional distribution.
3. No Need for Tuning Proposal Variance: Metropolis-Hastings requires tuning of the proposal variance, but slice sampling adjusts its step size dynamically.

- Initial Value: The Markov Chain starts at x = 0.25.
- Steps for Gibbs Sampling Using Slice Sampling:
1. Start with an sample y
2. Define the slice using y ~ $N(0,f(x_0)$
3. Find the Slice Interval: the interval $[0,f(x_0)]$.
4. Sample uniformly from this slice.
5. Repeat for all parameters iteratively in the Gibbs cycle until the distribution stabilizes.
